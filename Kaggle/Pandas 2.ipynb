```python
import pandas as pd
pd.set_option("display.max_rows", 7)
short = pd.read_csv('../simulations.csv')
short=short.drop(['shift index: 0->plots h 1->plots phi 2->plots S',
                 'Unnamed: 16'], axis=1)#dropping columns with awkward names
short.head()
```

<code>groupby()</code> method slices a dataframe acording the different values present in a column (or groups of colums). Each slice contains only rows matching each value

```python
short.groupby('length of domain').groups #returns a dictionary
```

```python
thing=short.groupby('length of domain') #groupby object
thing['length of domain'].describe()
```

```python
thing['length of domain'].count()
```

maximum resolution used in each of the 'length of domain' groups

```python
thing['neq'].max()
```

<code>value_counts()</code> method can be thought of as a shortcut of <code>df.groupby('col').col.count()</code>. Note however that <code>value_counts()</code> can *present the results in a different order*

```python
short['length of domain'].value_counts()
```

the <code>apply()</code> method gives **direct acces to the sliced dataframe and allows us to operate on it**. 

For example we can find the maximum substrate temperature for each group of lenght of domain in the dataframe

**Note that** `loc[]` and `iloc[]` **do not work with groupby objects**. Use `apply()` instead

```python
thing.apply(lambda df:df.loc[df['substrate temperature'].idxmax()])
```

grouping by multiple variables

```python
short.groupby(['kappa','delta']).apply(lambda df:df.Ca.mean())
```

Notice that the command `short.neq/short['length of domain']` does an element-wise division and returns the result in a new pandas series

### Examples with string data

```python
wines = pd.read_csv('wine_reviews.csv')
wines.head()
```

who are the most common reviewers in the dataset?

```python
wines.groupby('taster_twitter_handle').taster_twitter_handle.count()
```

What is the best wine I can buy for a given amount of money? 

```python
wines.groupby('price').points.max() #slice by price and maximize poins in each slice
```

What are the minimum and maximum prices for each 'variety' of wine?

```python
wines.groupby('variety').price.agg([min, max])#the parameters of aggregate are functions
# which operate 
```

Are there significant differences in the average scores assigned by the various reviewers?

```python
wines.groupby('taster_name').points.mean()
```

Most expensive wine varieties. The command below sorts wine prices grouped by 'variety' in descending order based on `min` values first, `max` second.

```python
wines.groupby('variety').price.agg([min, max]).sort_values(by=['min', 'max'], ascending=False)
```

What combination of countries and varieties are most common? 


```python
wines['n'] = 0 #creates a new column named 'n' and fills the column with zeroes
series=wines.groupby(['country', 'variety']).n.count().sort_values(ascending=False)
print(series)
```

```python
wines.groupby(['country', 'variety']).n.describe()
```
