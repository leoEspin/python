* count the number of outcomes for each player

````python
# my solution
chess_games.groupby('white_id').victory_status.value_counts().head(20)
````

````python
# solution in the tutorial:
(chess_games
    .assign(n=0)#adds column 'n' filled with zeroes
    .groupby(['white_id', 'victory_status'])
    .n
    .apply(len)#tells us the length of a given group!
    .reset_index()#removes the double index which results from groupby
    .head(20)
)
````

* Create a DataFrame like the one above, but only include users who are in the top 20 users by number of games played

````python
#display users by number of games played. notice use of groupby and len to count events
chess_games.groupby('white_id').apply(len).sort_values(ascending=False)[0:10]

#equivalent to
#chess_games['white_id'].value_counts()[0:10]
````

````python
# my solution:
users=chess_games['white_id'].value_counts()[0:20].index.tolist()
selected=chess_games.loc[chess_games['white_id'].apply(lambda x: True if x in users else False)]
(selected
    .assign(n=0)#adds column 'n' filled with zeroes
    .groupby(['white_id', 'victory_status'])
    .n
    .apply(len)#tells us the length of the group!
    .reset_index()
)
````

````python
# tutorial solution
(chess_games
    .assign(n=0)
    .groupby(['white_id', 'victory_status'])
    .n
    .apply(len)
    .reset_index()
    .pipe(lambda df: df.loc[df.white_id.isin(chess_games.white_id.value_counts().head(20).index)])
)
````

The `points` column in the `wine_reviews` dataset is measured on a 20-point scale between 80 and 100. Create a `Series` which normalizes the ratings so that they fit on a 1-to-5 scale instead (e.g. a score of 80 translates to 1 star, while a score of 100 is five stars). Set the `Series` name to "Wine Ratings", and sort by index value (ascending).

````python
#my solution:
(wines.points
     .apply(lambda x: x/5 -15)
     .value_counts()
     .rename('Wine Ratings')
     .sort_values(ascending=False)
     #.hist(bins=5)
)
````

solution in the tutorial: (notice that he sorts the index, not the values)

````python
# tutorial solution:
(((wines['points'].dropna() - 75) / 5)
    .value_counts()
    .sort_index()
    .rename_axis("Wine Ratings")
)
````

The `Stars` column in the `ramen_reviews` dataset is the ramen equivalent to the similar data points in `wine_reviews`. It is already on a 0-to-5 scale, but it has some different problems. Create a `Series` counting how many ramens earned each of the possible scores in the dataset. Convert the `Series` to the `float64` dtype and drop rames whose rating is `"Unrated"`. Set the name of the `Series` to "Ramen Ratings". Sort by index value (ascending).

** my solution:**

```python
#find indices for unrated rows first
inds=ramen_reviews.loc[ramen_reviews.loc[:,'Stars'] == 'Unrated'].index.values
(ramen_reviews
     .drop(inds)
     .Stars
     .astype('float64') #converts to float64 type
     .value_counts()
     .sort_index()
    .rename_axis("Ramen Ratings")
)```

** tutorial solution:**

```python
(ramen_reviews
    .Stars
    .replace('Unrated', None)#NOTE use of replace
    .dropna()
    .astype('float64')
    .value_counts()
    .rename_axis("Ramen Reviews")
    .sort_index())```
